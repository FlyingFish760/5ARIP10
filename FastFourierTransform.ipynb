```python
# Import necessary packages
import pywt
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, ifft
from scipy.io import wavfile
from pylab import *
```

```python
filename = r"C:\Users\piete\OneDrive\Documents\MSc\Q3\5ARIP10\DCASE_dataset\5ARIP10\dev_gearbox\gearbox\train\section_00_source_train_normal_0001_volt_2.5.wav"
samplerate, data = wavfile.read(filename)
a = data
b = a
c = fft(b) # calculate fourier transform (complex numbers list)
d = len(c)/2 
```
