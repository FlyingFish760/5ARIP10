```python
# Import necessary packages
import pywt
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, ifft
from scipy.io import wavfile
from pylab import *
import os
```

```python
training_data = r'C:\Users\piete\Documents\dev_gearbox\gearbox\train'

volt10 = []
samplerates = []
datas = []
# make a list of all the files in the training data folder that contain section_00 in the name and volt_1.0 in the name
for i in os.listdir(training_data):
    if 'section_00' in i and 'volt_1.0' in i:
        volt10.append(i)
        samplerate, data = wavfile.read(training_data + '\\' + i)
        samplerates.append(samplerate)
        datas.append(data)
# # check length of list
print(samplerate)
```

```python
FFTlengthhalf = []
FFT = []
for i in range(0, len(datas)):
    FFT.append(fft(datas[i]))
    FFTlengthhalf.append(len(FFT[i])/2)

print(FFT[1])
    
```

```python
# plot all the signals in datas
fig, ax = plt.subplots()
for i in range(0, len(datas)):
    ax.plot(datas[i])

plt.xlabel('Samples')
plt.ylabel('Amplitude for 1.0 volt')
plt.title('Signal Amplitude for 1.0 volt')
plt.show()
```

```python
# plot all the signals in datas
fig, ax = plt.subplots()
for i in range(0, len(datas)):
    ax.plot(abs(FFT[i][0:int(FFTlengthhalf[i])]), 'r')
    plt.xlim(0,30000)
    
plt.title('Fast fourier transform 1.0 volt')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()
```
